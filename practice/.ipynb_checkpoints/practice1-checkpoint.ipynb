# Import libraries

```python
import librosa
import librosa.display
import matplotlib.pyplot as plt
import IPython.display as ipd
import numpy as np
```

# Load audio files

```python
filename = "../audio/095_Coffee_House_10s.wav"

# 22050Hz is the default sampling rate: resamples the input to 22050Hz
# Mono is the default number or channel: mix stereo to mono 
#y, sr = librosa.load(filename) 
#y, sr = librosa.load(filename, sr=None) # use the native sampling rate
y, sr = librosa.load(filename, sr=None, mono=False) # use the native sampling rate and native channels

print ('sampling rate = ' + str(sr) + ' Hz')
print('audio length = ' +str(y.size)+ ' samples or ' + str(y.size/sr)+ 'sec' )



```

# Play audio files

```python
ipd.Audio(y, rate=sr) 
```

# Display waveforms

```python
plt.figure(figsize=(10, 5))
librosa.display.waveplot(y, sr=sr)
```

# Compute spectrogram

```python
hop_length = 512
n_fft = 2048
Y = librosa.stft(y, n_fft=n_fft, hop_length=hop_length)
D = librosa.amplitude_to_db(np.abs(Y), ref=np.max)

#D = librosa.amplitude_to_db(np.abs(librosa.stft(y)), ref=np.max)
```

# Display Spectrogram

```python
#fig, ax = plt.subplots(nrows=1, ncols=1, sharex=True)

plt.figure(figsize=(10, 5))

img = librosa.display.specshow(D, y_axis='linear', x_axis='time', hop_length=hop_length, sr=sr)
plt.colorbar(format='%+2.0f dB')

#ax[0].set(title='Linear-frequency power spectrogram')
#ax[0].label_outer()
```

```python
#filename = "audio/095_Coffee_House_10s.wav"
filename = "audio/piano_88keys_short_vel96_part2.wav"

# sr=22050 is the default sampling rate. The original sampling rate is resampled to 22050 Hz 
y, sr = librosa.load(filename)   
print(y.size)
print (sr)


hop_length = 512
n_fft = 2048
Y = librosa.stft(y, n_fft=n_fft, hop_length=hop_length)
D = librosa.amplitude_to_db(np.abs(Y), ref=np.max)

plt.figure(figsize=(10, 5))

img = librosa.display.specshow(D, y_axis='linear', x_axis='time', hop_length=hop_length, sr=sr)
plt.colorbar(format='%+2.0f dB')

#ax[0].set(title='Linear-frequency power spectrogram')
#ax[0].label_outer()
```

```python
hop_length = 512
n_fft = 2048

D = np.abs(librosa.stft(y, n_fft=n_fft, hop_length=hop_length))**2
S = librosa.feature.melspectrogram(S=D, sr=sr)


#Y = librosa.stft(y, n_fft=n_fft, hop_length=hop_length)
#D = librosa.amplitude_to_db(np.abs(Y), ref=np.max)

plt.figure(figsize=(10, 5))


#fig, ax = plt.subplots()
S_dB = librosa.power_to_db(S, ref=np.max)
img = librosa.display.specshow(S_dB, x_axis='time',
                         y_axis='mel', sr=sr,
                         fmax=8000)
plt.colorbar(format='%+2.0f dB')
#ax.set(title='Mel-frequency spectrogram')


#img = librosa.display.specshow(D, y_axis='linear', x_axis='time', hop_length=hop_length, sr=sr)
#plt.colorbar(format='%+2.0f dB')

#ax[0].set(title='Linear-frequency power spectrogram')
#ax[0].label_outer()
```

```python
#y, sr = librosa.load(librosa.ex('trumpet'))
C = np.abs(librosa.cqt(y, sr=sr, fmin=librosa.note_to_hz('C2')))
#fig, ax = plt.subplots()


plt.figure(figsize=(10, 5))
img = librosa.display.specshow(librosa.amplitude_to_db(C, ref=np.max),
                               sr=sr, x_axis='time', y_axis='cqt_note', fmin=librosa.note_to_hz('C2'))
#ax.set_title('Constant-Q power spectrum')
plt.colorbar(img, format="%+2.0f dB")


```
